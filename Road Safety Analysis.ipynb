# EDA - Main Attributes of Accidents Dataset

```python
# Has the number of accidents increased or decreased over the last few months?---------------------------

sns.set_style('white')
fig, ax = plt.subplots(figsize=(15,6))

accidents.set_index('date').resample('M').size().plot(label='Total Accidents per Month in 2020', color='grey', ax=ax)

ax.set_title('Accidents per Month in 2020', fontsize=14, fontweight='bold')
ax.set(ylabel='Total Count\n', xlabel='')
ax.legend(bbox_to_anchor=(1.1, 1.1), frameon=False)

sns.despine(ax=ax, top=True, right=True, left=True, bottom=False);
```

```python
# On which weekdays are accidents most likely to occur?---------------------------

weekday_counts = pd.DataFrame(accidents.set_index('date').resample('1d')['accident_index'].size().reset_index())
weekday_counts.columns = ['date', 'count']

weekday = weekday_counts['date'].dt.weekday_name

weekday_averages = pd.DataFrame(weekday_counts.groupby(weekday)['count'].mean().reset_index())
weekday_averages.columns = ['weekday', 'average_accidents']
weekday_averages.set_index('weekday', inplace=True)
weekday_averages
```

```python
days = ['Sunday', 'Saturday', 'Friday', 'Thursday', 'Wednesday', 'Tuesday', 'Monday']

sns.set_style('white')
fig, ax = plt.subplots(figsize=(10,5))
colors=['indigo', 'indigo', 'indigo', 'darkgoldenrod', 
        'indigo', 'indigo', 'indigo']

weekday_averages.reindex(days).plot(kind='barh', ax=ax, color=[colors])
ax.set_title('\nAverage Accidents per Weekday\n', fontsize=14, fontweight='bold')
ax.set(xlabel='\nAverage Number', ylabel='')
ax.legend('')

sns.despine(ax=ax, top=True, right=True, left=True, bottom=True);
```

```python
# What was the severity of the accidents?---------------------------

accidents.accident_severity.value_counts()

```

```python
fatal   = accidents.accident_severity.value_counts()['1 - Fatal']
serious = accidents.accident_severity.value_counts()['2 - Serious']
slight  = accidents.accident_severity.value_counts()['3 - Slight']

names = ['Fatal Accidents','Serious Accidents', 'Slight Accidents']
size  = [fatal, serious, slight]

plt.pie(x=size, labels=names, colors=['darkgoldenrod', 'indigo', 'silver'], 
        autopct='%1.2f%%', pctdistance=0.6, textprops=dict(fontweight='bold'),
        wedgeprops={'linewidth':7, 'edgecolor':'white'})

my_circle = plt.Circle((0,0), 0.6, color='white')

fig = plt.gcf()
fig.set_size_inches(8,8)
fig.gca().add_artist(my_circle)
plt.title('\nAccident Severity Percentages (2020)', fontsize=14, fontweight='bold')
plt.show()
```

```python
# How are accidents distributed throughout the day? -----------------

sns.set_style('white')
fig, ax = plt.subplots(figsize=(10,6))

accidents.hour.hist(bins=24, ax=ax, color='indigo')
ax.set_title('\nAccidents depending by Hour of the Day\n', fontsize=14, fontweight='bold')
ax.set(xlabel='Hour of the Day', ylabel='Total Count of Accidents')

sns.despine(top=True, right=True, left=True, bottom=True);
```

```python
# prepare dataframe with simple counts
counts = accidents.groupby(['daytime', 'accident_severity']).size()

counts = counts.rename_axis(['daytime', 'accident_severity'])\
                                .unstack('accident_severity')\
                                .rename({1:'fatal', 2:'serious', 3:'slight'}, axis='columns')
counts
```

```python
male   = vehicles.sex_of_driver.value_counts()['Male']
female = vehicles.sex_of_driver.value_counts()['Female']
unknown = vehicles.sex_of_driver.value_counts()['Unknown']
missing = vehicles.sex_of_driver.value_counts()['Missing data or out of range']

names = ['Male','Female', 'Unknown', 'Missing']
size  = [male, female, unknown, missing]

plt.pie(x=size, labels=names, colors=['indigo', 'darkgoldenrod', 'silver'], 
        autopct='%1.2f%%', pctdistance=0.6, textprops=dict(fontweight='bold'),
        wedgeprops={'linewidth':7, 'edgecolor':'white'})

my_circle = plt.Circle((0,0), 0.6, color='white')

fig = plt.gcf()
fig.set_size_inches(8,8)
fig.gca().add_artist(my_circle)
plt.title('\nSex of Drivers (2020)', fontsize=14, fontweight='bold')
plt.show()
```

```python

drivers = vehicles.groupby(['age_band_of_driver', 'sex_of_driver']).size().reset_index()

drivers.drop(drivers[(drivers['age_band_of_driver'] == '0') | \
                     (drivers['sex_of_driver'] == '3') | \
                     (drivers['sex_of_driver'] == '-1')]\
             .index, axis=0, inplace=True)

drivers.columns = ['age_band_of_driver', 'sex_of_driver', 'count']
drivers
```

```python

fig, ax = plt.subplots(figsize=(14, 7))
sns.barplot(y='age_band_of_driver', x='count', hue='sex_of_driver', data=drivers, palette='magma')
ax.set_title('\nAccidents\' Drivers by Age and Sex\n', fontsize=14, fontweight='bold')
ax.set(xlabel='count', ylabel='age band of driver')
ax.legend(bbox_to_anchor=(1.1, 1.), borderaxespad=0., frameon=False)

sns.despine(top=True, right=True, left=True, bottom=True);
```
